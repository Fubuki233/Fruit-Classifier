# Category Static

```python
import os
import matplotlib.pyplot as plt

data_dir = "data/train"

# 只保留真正的目录
classes = [c for c in os.listdir(data_dir) 
           if os.path.isdir(os.path.join(data_dir, c))]

class_counts = {
    cls: len([f for f in os.listdir(os.path.join(data_dir, cls)) 
              if f.endswith('.jpg')])
    for cls in classes
}

# ---- 绘图 ----
plt.figure(figsize=(8, 5))
bars = plt.bar(class_counts.keys(), class_counts.values(), color='skyblue')

plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Images")

# ---- 在柱状图顶部标数字 ----
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # x 坐标
        height,                              # y 坐标（柱子顶部）
        f"{int(height)}",                    # 显示数量
        ha='center', va='bottom', fontsize=10
    )

plt.tight_layout()
plt.show()
```
