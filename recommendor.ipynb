```python
import pandas as pd
import numpy as np
df = pd.read_csv('D:/Nus/Fruit-Classifier/mini_item_based.csv')
df.head()
```

```python

year = df["Year"].values
year =  (year - np.min(year))/(np.max(year) - np.min(year))
#year[:5]
dim = df["Dur_In_Min"].values
dim =  (dim - np.min(dim))/(np.max(dim) - np.min(dim))
#dim[:5]
df["Year"] = year
df["Dur_In_Min"] = dim

df.head()
```

```python
from scipy import spatial

user_row = df[df["Movie_Name"] == "m5"].iloc[0]

othermovies = df[df["Movie_Name"] != "m5"]

feature_cols = ["Year", "Dur_In_Min","English","French"] 

user_vec = user_row[feature_cols].values.astype(float)

candidates = []  # (distance, movie_name)

for _, movie in othermovies.iterrows():
    movie_vec = movie[feature_cols].values.astype(float)
    distance = spatial.distance.cosine(user_vec, movie_vec)
    candidates.append((distance, movie["Movie_Name"]))

top2 = sorted(candidates, key=lambda x: x[0])[:2]

for dist, name in top2:
    print(f"{name}, cosine dist = {dist:.4f}")
```
